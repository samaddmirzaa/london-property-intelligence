### Data Profiling

#### Importing Libraries

```python
import numpy as np
import pandas as pd
from pathlib import Path
```

#### 33 Borough Files

```python
# Loading the 33 London Borough Housing Files

project_root = Path('.').resolve().parent
raw_dir = project_root /'data'/'raw'
borough_files = sorted(raw_dir.glob('*_link_*.csv'))
print(f'Found {len(borough_files)} borough files')
for fp in borough_files:
    print(' ', fp.name)
```

```python
# Merging into Single Data Frame

df = pd.concat(
    [pd.read_csv(fp) for fp in borough_files],
    ignore_index=True
)
df.head()
```
